```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('darkgrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

##### plot which represents correlation between fare and age

```python
sns.jointplot(x='fare',y='age',data=titanic)
```

##### plot which represents correlation between age and wealthiness (class)

```python
sns.boxplot(x='class',y='age',data=titanic,palette='viridis',hue='who')
```

##### same as above, but using swarmplot and hue to sex

```python
sns.swarmplot(x='class',y='age',data=titanic,hue='sex',dodge=True,palette='cubehelix')
```

##### plot which represents where male and female start their travel

```python
sns.countplot(x='sex',data=titanic,hue='embark_town')
```

##### matrix which represents correlations of columns

```python
titanic.corr()
```

```python
sns.heatmap(titanic.corr(),cmap='coolwarm',square=True)
```

##### FacetGrid which represents density of age split for male and female

```python
sns.FacetGrid(data=titanic,col='sex').map(sns.distplot,'age')
```
