```python
import numpy as np
x = np.arange(0,11)
y = x*2
y2 = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

##### drawing a simple y = 2x plot

```python
fig = plt.figure(figsize=(3,3),dpi=100)
ax = fig.add_axes([0,0,0.5,1])
ax.plot(x,y,'g--')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('y=2x')
```

##### drawing two y = 2x plots, with different parameters

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([1.2,0.5,.6,.6])

ax1.plot(x,y2,x,y,label='x**2',ls='--',marker='o',lw=2)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('y=x^2')
ax1.axis('tight')
ax1.legend(loc=0)

ax2.plot(x,y2,x,y,label='2*x',ls='-',marker='s',lw=1,markersize=10,markerfacecolor='#40A948',markeredgewidth=2,markeredgecolor='#FF0000')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('y=2x')
ax2.set_xlim(-0.1,4.1)
ax2.set_ylim(-1,21)
ax2.legend(loc=9)
```

#### drawing plots using plt.subplots()

```python
fig, ax = plt.subplots(nrows=1,ncols=2)

ax[0].plot(x,y2)
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')
ax[0].set_title('y=x^2')

ax[1].plot(x,y)
ax[1].set_xlabel('X')
ax[1].set_ylabel('Y')
ax[1].set_title('y=2x')
```
